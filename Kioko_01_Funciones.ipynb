# Funci√≥n lineal


Tiene la forma de: ùëì(ùë•)=ùëöùë•+ùëè
 
donde  ùëö  y  ùëè   ‚ààùëÖ .

ùëö  puede ser calculada por:
ùëö=(ùë¶2‚àíùë¶1) / (ùë•2‚àíùë•1) y  ùëè, es el punto de corte con el eje  ùë¶.

Su dominio es  ùê∑ùëúùëöùëì=(‚àí‚àû,‚àû) . Su Rango es  ùêºùëöùëì=(‚àí‚àû,‚àû)

```python
resolucion = 100 #n√∫mero de puntos de la receta .. Este valor no puede ser flotante... Hay que saber cuando asignar un valor flotante y cuando no...
#En computaci√≥n todos los valores son discretos.
```

```python
def f( m, x, b ): #Funciones y ciclos  necesitamos los dos puntitos para definir que ah√≠ empieza
  return m * x + b
```

```python
#Puntos (2,6) y (3,7)
#      ( x1, y1 ) y ( x2 , y2)
m=( 7. - 6. )/( 3. - 2. )
```

```python
x = np.linspace( -10. , 10. , resolucion) #Me genera mi vector, yo defino mi inicio y mi fin (rango) y cuantos puntos quiero que me de.
#x
corte_y = 9
y = f(m , x , corte_y) #No importa que "b" aqu√≠ se llame "corte_y" , el chiste es respetar el orden de como se vana  ir llamando mis variables.
y2 = f(corte_y , x , m) #El orden es importante
plt.plot( x , y , label = '1' ) 
plt.plot( x , y2 , label = '2' )
plt.legend() #Instruccion para que aparezcan las etiquetas.
```

# Composici√≥n de Funciones

```python
def g(x):
  return x**2. #En python potencia es dos veces * --> **
  #Todo lo identado, Python  toma ltodo lo que est√° adentro como parte de la funci√≥n.

def f(x):
  return np.sin(x)

valores_de_x = np.linspace( -5. , 5. , 1000 )

f_o_g = f(g(valores_de_x)) #f_o_g = sen(x^2)
plt.plot(valores_de_x , f_o_g)
plt.plot(valores_de_x , f(valores_de_x))
plt.plot(valores_de_x , g(valores_de_x))
plt.ylim(-2,4) #definimos el l√≠mite que deseamos ver en y
plt.xlim(-4,4) #definimos el l√≠mite que deseamos ver en x
```

# Actividad 4: Grafiquen la funci√≥n f(g(x)), siendo f(x)=3x+2 y, g(x)=2x

```python
def f (x):
  return 3*x + 2

def g(x):
  return 2*x

f_o_g = f(g(x)) #Funcion 6x + 2
x = np.linspace(-10 , 10 , 100)
plt.plot(x , f_o_g)
plt.grid()
```

# Actividad 5: Grafiquen la funci√≥n f(g(x)), siendo f(x)=sin(x) y, g(x)=1/x

```python
def f (x):
  return np.sin(x)

def g(x):
  return 1/x

f_o_g = f(g(x)) #Funcion sin(1/x)
x = np.linspace(-1. , 1. , 100)
plt.plot(x , f_o_g)
```

# Actividad 6: Grafiquen la funci√≥n f(g(x)), siendo f(x)=2x+1 y, g(x)=cos(x)

```python
def f (x):
  return 2*x + 1

def g(x):
  return np.cos(x)

f_o_g = f(g(x)) #Funcion 2*cos(x) + 1
x = np.linspace(-10. , 10. , 100)
plt.plot(x , f_o_g)
```
